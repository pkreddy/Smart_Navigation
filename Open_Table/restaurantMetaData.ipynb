```python
import urllib.request as url
import pandas as pd
import traceback
from bs4 import BeautifulSoup
from lxml import html
open_table_url = "https://m.opentable.com"
```

```python
def getUrlData(url_string):
    try:
        #print("\nreading the response to the url -" + url_string)
        response = url.urlopen(url_string)
        # print(response.read())
        soup = BeautifulSoup(response, 'html.parser')
        # print(soup.prettify())
        tree = html.fromstring(str(soup))
        return tree
    except Exception as inst:
        #traceback.print_exc()
        print("Excpetion in getUrlData for url - ",url_string)
        print(inst)
        return ""
```

```python
def getMetaData(meta_url):
    try:
        returnMetaData = list()
        response = getUrlData(meta_url)
        dining_style = (response.xpath("//span[text()='Dining Style:']/..//span[2]/text()"))
        cuisines = (response.xpath("//span[text()='Cuisines:']/..//span[2]/text()"))
        hours_of_operation = (response.xpath("//span[text()='Hours of Operation:']/..//span[2]/text()"))
        price_range = (response.xpath("//span[text()='Price Range:']/..//span[2]/text()"))
        returnMetaData = [''.join(dining_style),''.join(cuisines),''.join(price_range),''.join(hours_of_operation)]
        return(returnMetaData)
    except Exception as inst:
        print("Exception in getMetaData for url -",meta_url)
        print(inst)
        returnMetaData = ['','','','']
        return(returnMetaData)
```

```python
def getMenu(menus_url):
    try:
        Menu_List = ""
        response = getUrlData(menus_url)
        menu = response.xpath("//a/@href")
        # //ul[@class='menu-category__slider']/li
        for i in range(len(menu)):
            if('menu' in str(menu[i])):
                response_menu = getUrlData(open_table_url+str(menu[i]))
                Menu_List = Menu_List + ";" +';'.join(response_menu.xpath("//span[@class='menu-details__col--10']/text()"))
        return Menu_List
    except Exception as inst:
        print("Exception in getMenu")
        print(inst)
        return ('')
```

```python
def getRatingsAndReviews(ratings_url):
    try:
        RatingAndReviews = list()
        response = getUrlData(ratings_url)
        rating = response.xpath("//h3[text()='Overall Rating']/..//div/text()")
        overall_rating = str(rating[0])
        recommended_to_a_friend = str(rating[1].split("%")[0])
        food = str(response.xpath("//dt[text()='Food']/..//dd/text()")[0])
        service = str(response.xpath("//dt[text()='Service']/..//dd/text()")[0])
        ambiance = str(response.xpath("//dt[text()='Ambiance']/..//dd/text()")[0])
        noise_level = str(response.xpath("//dt[text()='Noise Level']/..//dd/text()")[0])
        RatingAndReviews = [overall_rating,recommended_to_a_friend,food,service,ambiance,noise_level]
        return (RatingAndReviews)
        #print(rating[0],rating[1].split("%")[0])
    except Exception as inst:
        print("Exception in getRatingsAndReviews")
        print(inst)
        RatingAndReviews = ['','','','','']
        return(RatingAndReviews)
```

```python
xml = ['<restaurants>']
df1 = pd.DataFrame(columns=["id","overall_rating","recommended_to_a_friend","food","service","ambiance","noise_level"
,"dining_style","cuisines","price_range","hours_of_operation"])
table_data = pd.read_csv('open_table_data_final.csv',encoding = "ISO-8859-1")
for i in range(0,len(table_data)):
    try:
        id = str(table_data.at[i,"id"])
        xml.append('<menu id=\"'+id+'\">')
        response = getUrlData(table_data.at[i,"mobile_reserve_url"])
        a = getRatingsAndReviews(open_table_url+(response.xpath("//a[@id='reviewsLink']/@href"))[0])
        c = str(getMenu(open_table_url+(response.xpath("//a[@id='menuLink']/@href"))[0]))
        b = getMetaData(table_data.at[i,"reserve_url"])
        data = [id,*a,*b]
        xml.append(c.replace(",",""))
        xml.append('</menu>')
        df1.loc[i] = data
        #print(data)
    except Exception as inst:
        print(inst)
        print(id)
        df1.loc[i] = [id,"","","","","","","","","",""]
xml.append('</restaurants>')
```
